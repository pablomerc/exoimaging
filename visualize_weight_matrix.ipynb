```python
import torch
import numpy as np
from weights_matrix import simulate_circle_movement
import os
from config import (
    MATRIX_SIZE,
    NUM_TIMESTEPS,
    CIRCLE_RADIUS,
    OUTPUT_DIR,
    LIGHTCURVE_PATH,
    WEIGHTS_MATRIX_PATH,
    RECOVERED_IMAGE_PATH,
    RECOVERED_IMAGE_PLOT_PATH,
)
import matplotlib.pyplot as plt

from weights_matrix import simulate_circle_movement

matrix_size=34
t=500
_,weights_matrix = simulate_circle_movement(
    matrix_size=matrix_size,
    num_timesteps=t,
    circle_radius=30,
    equator_shift=-15
)

weights_matrix = weights_matrix.reshape((t,matrix_size**2))

U, s, Vt = np.linalg.svd(weights_matrix)

plt.figure(figsize=(10, 6))
plt.semilogy(s, 'o-')
plt.title('Singular Value Spectrum')
plt.xlabel('Singular Value Index')
plt.ylabel('Singular Value (log scale)')
plt.grid(True)
plt.show()
```

```python
plt.imshow(weights_matrix.T, cmap='gray', interpolation='nearest')
plt.colorbar(label='Intensity')
plt.title('100100 Matrix (Grayscale)')
plt.xlabel('Spatial Position (flattened 1010 grid)')
plt.ylabel('Time (timesteps)')
plt.show()
```
